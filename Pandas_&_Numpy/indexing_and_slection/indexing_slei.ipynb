```python
import pandas as pd 
import numpy as np 
from pathlib import Path 
```

```python
# call files and reading 
file_path = Path("ramen-ratings.csv")

data =pd.read_csv(file_path)
data.head(10) # the firs 10 rows 
```

```python
# select the country from columns 
print(data['Country'])
print('all Country names in this columns \n','*'*30)

print(data['Country'].unique())
print('\n','*'*30,'\njust unique country names')
print(data['Country'].tolist())
print('Country in List python ')
```
