#Assing Data

```python
try :
    data['Top Ten']=data['Top Ten'].astype(int)
except ValueError :
      data[['Year','Rank']]=data['Top Ten'] = data['Top Ten'].str.extract(r'(\d{4})\s+#(\d)').astype(int)
```

```python
#drop 

# Extract the year and ranking from Top Ten
df =data.loc[data['Top Ten'].notnull()]
df = df.copy()

extracted = df['Top Ten'].str.extract(r'(\d{4})\s+#(\d+)')  #filter [2016#10] \d(4 number) plus 1 \d(1 number) //first and seconf columens



#insert 2 new columens in the DataFrame 
df.loc[:,'Year'] = pd.to_numeric(extracted[0],errors ='coerce').astype('Int64') # str to int //Int suport Nan Value big I but int not
                                 # extracted[0] the first columns (\d{4}) == Year
df.loc[:,'Rank'] = pd.to_numeric(extracted[1],errors='coerce').astype('Int64')
                                 # extracted[1] the second columns (\d+) == Rank  
df.loc[:,'Year'] = df['Year'].ffill().bfill() # return value from last line
df.loc[:,'Rank'] = df['Rank'].fillna(5) # if Ranke =0 return 5 
# df.drop(columns=['Year', 'Rank'], errors='ignore', inplace=True)

df
```
